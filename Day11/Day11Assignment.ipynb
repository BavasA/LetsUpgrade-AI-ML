```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
dataset1=pd.read_csv("general_data.csv")
dataset1.head()
```

```python
dataset1["Attrition"] = dataset1["Attrition"].map({"Yes":1,"No":0})
dataset1.head()
```

```python
from scipy.stats import pearsonr
statst,p = pearsonr(dataset1.Age,dataset1.Attrition)
print(statst)
print(p)
print("{:f}".format(p))
```

# Correlation1:
r value is not near to 1 or -1 so Age is not factor for attrition

```python
statst,p = pearsonr(dataset1.DistanceFromHome,dataset1.Attrition)
print(statst)
print(p)
print("{:f}".format(p))
```

# Correlation2:
r value is not near to 1 or -1 so DistanceFromHome is not a major factor for attritiom

```python
statst,p = pearsonr(dataset1.MonthlyIncome,dataset1.Attrition)
print(statst)
print(p)
print("{:f}".format(p))
```
