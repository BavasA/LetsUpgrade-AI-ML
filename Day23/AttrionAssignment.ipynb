```python
import pandas as pd 
import numpy as np 
from sklearn import tree
from sklearn import preprocessing
```

```python
data = pd.read_csv("general_data.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.NumCompaniesWorked = np.where(data.NumCompaniesWorked.isnull(),
                                  data.NumCompaniesWorked.mean(),
                                  data.NumCompaniesWorked)
```

```python
data.TotalWorkingYears = np.where(data.TotalWorkingYears.isnull(),
                                  data.TotalWorkingYears.mean(),
                                  data.TotalWorkingYears)
```

```python
data.isnull().sum().sum()
```

```python
from sklearn import preprocessing
# le = preprocessing.LabelEncoder()
encoded_data = data.apply(preprocessing.LabelEncoder().fit_transform)
encoded_data
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
rf_model
```

```python
data.columns
features=['Age','BusinessTravel', 'Department', 'DistanceFromHome','Education', 'EducationField',
          'Gender','JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome','NumCompaniesWorked',
          'PercentSalaryHike','TotalWorkingYears','YearsAtCompany','YearsSinceLastPromotion', 
          'YearsWithCurrManager']
```

```python
rf_model.fit(X=encoded_data[features],y=encoded_data["Attrition"])
```

```python
print("OOB Accurary")
print(rf_model.oob_score_)
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

```python
predictors = pd.DataFrame(encoded_data["YearsAtCompany"],encoded_data["Department"])
                          
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth=8)
tree_model.fit(X=predictors,y=encoded_data["Attrition"])
```

```python
with open("Dtree8.dot","w")as f:
    f=tree.export_graphviz(tree_model,feature_names=predictors.columns,out_file=f)
```

```python
tree_model.score(X=predictors,y=encoded_data["Attrition"])
```
